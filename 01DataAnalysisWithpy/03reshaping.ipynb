```python
import numpy as np

arr = np.arange(6)
print(arr)
```

```python
reshaped = arr.reshape((2,3))
print(reshaped)
```

```python
#flatten
flattended = reshaped.flatten()
print(flattended)
```

```python
#transpose 
transpose = reshaped.T
print(f"reshaped :\n{reshaped}")
print(f"transpose :\n{transpose}")
```
