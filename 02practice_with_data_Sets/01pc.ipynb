```python
import numpy as np

restaurants = np.array([
    [101, 120, 130, 140, 150, 160],
    [102,  90,  95, 100, 110, 120],
    [103, 200, 210, 220, 230, 250],
    [104,  60,  70,  75,  80,  85],
    [105, 150, 155, 160, 165, 170]
])

print(restaurants)
```

```python
#matplot lib

import matplotlib as plt

print(restaurants.shape)
print(f"for first 3")
print(restaurants[:3])
print(f"from 3rd till end")
print(restaurants[3:])

```

```python
# total sales per year
print(np.sum(restaurants,axis=0))
print(np.sum(restaurants,axis=0)[1:])

#yearly total
yearly_total = np.sum(restaurants[:,1:],axis=0)
print(yearly_total)

col_total = np.sum(restaurants[:,1:],axis=1)
print(col_total)
```

```python
min_sales0 = np.min(restaurants[:,1:], axis=0)
min_sales1 = np.min(restaurants[:,1:], axis=1)

print(min_sales0)
print(min_sales1)
```

```python
max_sales0 = np.max(restaurants[:,1:], axis=0)
max_sales1 = np.max(restaurants[:,1:], axis=1)

print(max_sales0)
print(max_sales1)
```

```python
#avergae 

avg_sales0 = np.average(restaurants[:,1:], axis=0)
avg_sales1 = np.average(restaurants[:,1:], axis=1)

print(avg_sales0)
print(avg_sales1)
```

```python
#cumulative sum
cumsum = np.cumsum(restaurants[:,1:], axis=1)
print(cumsum)
```

```python
import matplotlib.pyplot as plt

figure = plt.figure(figsize=(5,5))
plt.plot(np.mean(cumsum,axis=0))
plt.title("avg cum sales")
plt.xlabel("years")
plt.ylabel("sales")
plt.grid(True)
plt.show()
```
